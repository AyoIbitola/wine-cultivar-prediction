# Wine Cultivar Prediction - Model Development

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
```

## 2. Load Dataset
Loading the Wine dataset from sklearn.

```python
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

## 3. Data Preprocessing
We need to select 6 specific features as per the project requirements.
Selected features: 
1. `alcohol`
2. `flavanoids`
3. `color_intensity`
4. `hue`
5. `proline`
6. `od280/od315_of_diluted_wines`

```python
selected_features = ['alcohol', 'flavanoids', 'color_intensity', 'hue', 'proline', 'od280/od315_of_diluted_wines']
X = df[selected_features]
y = df['target']

# Check for missing values
print(X.isnull().sum())
```

### Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## 4. Model Training
Using Random Forest Classifier.

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

## 5. Evaluation

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

## 6. Save Model

```python
joblib.dump(model, 'wine_cultivar_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Model and scaler saved successfully.")
```
